```python
import requests
import json
import pandas as pd
```

```python
query = """query trades{
  pairs(first: 200,orderby: txCount, orderDirection: desc) {
    token0{
      name,
      tradeVolumeUSD,
      totalLiquidity

    }
    token1{
      
      name,
      tradeVolumeUSD,
      totalLiquidity

    },

    token0Price,
    token1Price,
    volumeUSD,
    txCount,
    
    createdAtTimestamp,


  }
 
}"""
```

```python
url = 'https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2'
r = requests.post(url, json={'query': query})
print(r.status_code)
print(r.text)
```

```python
json_data = json.loads(r.text)
json_data
```

```python
df_data = json_data["data"]["pairs"]
df = pd.DataFrame(df_data)
df
```

```python
df['token0']
```

```python
token0Name=[]
token0tradeVolumeUSD=[]
token0totalLiquidity=[]
for i in range(0,200):
  token0Name.append(df['token0'][i]['name'])
  token0tradeVolumeUSD.append(df['token0'][i]['tradeVolumeUSD'])
  token0totalLiquidity.append(df['token0'][i]['totalLiquidity'])


```

```python
token1Name=[]
token1tradeVolumeUSD=[]
token1totalLiquidity=[]
for i in range(0,200):
  token1Name.append(df['token1'][i]['name'])
  token1tradeVolumeUSD.append(df['token1'][i]['tradeVolumeUSD'])
  token1totalLiquidity.append(df['token1'][i]['totalLiquidity'])
```

```python
import numpy as np
token0Name= np.array(token0Name)
token0tradeVolumeUSD= np.array(token0tradeVolumeUSD)
token0totalLiquidity= np.array(token0totalLiquidity)
token1Name=np.array(token1Name)
token1tradeVolumeUSD=np.array(token1tradeVolumeUSD)
token1totalLiquidity=np.array(token1totalLiquidity)


```

```python
df1 = pd.DataFrame(token0Name, columns =['Token0 Name'])
df2 = pd.DataFrame(token1Name, columns =['Token1 Name'])
df3 = pd.DataFrame(token0tradeVolumeUSD, columns =['Token0 Trade Volume(USD)'])
df4 = pd.DataFrame(token1tradeVolumeUSD, columns =['Token1 Trade Volume(USD)'])
df5 = pd.DataFrame(token0totalLiquidity, columns =['Token0 Liquidity'])
df6 = pd.DataFrame(token1totalLiquidity, columns =['Token1 Liquidity'])

```

```python
frames = [df1, df2, df3, df4, df5, df6]
 
data = pd.concat(frames, axis=1)
data
```

```python
df= df.drop(['token0','token1'],axis=1)
df
```

```python
final_data= pd.concat([df,data],axis=1)
final_data
```
